```python
# Working with more than 2D arrays
# for example
import skimage
image_rgb = skimage.io.imread('https://github.com/guiwitz/PyImageCourse_beginner/raw/master/images/46658_784_B12_1.tif')
```

```python
# 3D image
image_rgb.shape
```

```python
# the above image has multiple channels
fig, ax = plt.subplots(figsize=(2.5, 2.5))
ax.imshow(image_rgb);
```

```python
# splitting the image into its individual channels
im_channel1 = image_rgb[:, :, 0]
im_channel2 = image_rgb[:, :, 1]
im_channel3 = image_rgb[:, :, 2]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))  

ax[0].imshow(im_channel1) 
ax[0].set_title('Channel 1')  

ax[1].imshow(im_channel2) 
ax[1].set_title('Channel 2')  

ax[2].imshow(im_channel3)  
ax[2].set_title('Channel 3') 
```
